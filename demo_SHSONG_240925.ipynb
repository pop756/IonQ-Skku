```python
from TBM_ver3 import TightBindingModel_Hamiltonian, TightBindingModel_VQE
import warnings

warnings.filterwarnings('ignore')
```

```python
######################
# 1-D monatomic chain #
######################

num_q_orbital=2
num_q_site=5

t_0=[[0,3],
    [3,0]]
t_1=[[0, 0],
    [3, 0]]

obj_hamiltonian=TightBindingModel_Hamiltonian(
    num_q_orbital=num_q_orbital,
    num_q_site=num_q_site,
    t_0=t_0,
    t_1=t_1
)
hamiltonian_dictionary=obj_hamiltonian.Hamiltonian()

obj_vqe=TightBindingModel_VQE(
    num_q_orbital=num_q_orbital,
    num_q_site=num_q_site,
    hamiltonian_dictionary=hamiltonian_dictionary,
    simulation_mode='noisy'
)

obj_vqe.Energy_Momentum_Dispersion(upper_bound_level=1)
```
