```python
from TBM_ver1 import TightBindingModel_Hamiltonian, TightBindingModel_VQE
from qiskit.primitives import StatevectorEstimator
import matplotlib.pyplot as plt

#########################
# Initialize conditions #
#########################

num_q_orbital=2
num_q_site=5
t_0=[[-0.5,1],
    [1,0.5]]
t_1=[[0, 1],
    [0, 0]]

obj_hamiltonian=TightBindingModel_Hamiltonian(
    num_q_orbital=num_q_orbital,
    num_q_site=num_q_site,
    t_0=t_0,
    t_1=t_1
)
hamiltonian=obj_hamiltonian.Hamiltonian()

obj_vqe=TightBindingModel_VQE(
    num_q_orbital=num_q_orbital,
    num_q_site=num_q_site,
    hamiltonian=hamiltonian
)

# eigen energy list : 1 dimensional list [row : eigen energy for sequential p's(0,1,...,2**n-1)]
eigen_energy=list()

# set Hamiltonian
A_N=obj_hamiltonian.get_A_N()
B_N=obj_hamiltonian.get_B_N()


# entire ansatze for 0th level and optimization with the for loop containing site ansatze and its corresponding quantum number p
for p in range(2**num_q_site):
    
    # site ansatze for 0th level
    qc_site=obj_vqe.ansatze_site(p)
    
    estimator = StatevectorEstimator()
    job = estimator.run([(qc_site, A_N)])
    result = job.result()[0]
    
    eigen_energy.append(result.data.evs)
    
x=range(2**num_q_site)
y=eigen_energy

plt.figure(figsize=(5, 3))
plt.plot(x, eigen_energy, 'bo-', markersize=2)

plt.xlabel('p')
plt.ylabel('Energy (E)')
plt.show()
```
